### **PANDAS** - contains high level data structure that aids faster data analysis and easier manipulations. Pandas is built on top of NumPy making it completely usable under NymPy Centric Applications.  


---


### The explicit features of Pandas are: 
-> Data structure with axes that suuports automatic and explicit data access and alignment.

-> Data Structures intergrated with time series functionalities.

-> Same Data Structure to handle time series and non-time series data.

-> Arithmetic and Logical operations passing iver the axis labels or the metadata of the axes.

-> Flexibility in handling data

-> Integrated Relational Functionalities that are common in RDBMS like merge,etc.


---

### The two types of Data Structures facilitated by Pandas :
▶ Series - Data Structure that handles 1-dimensional array-like objects with an associated array of labels, usually called the index.

▶ DataFrame - Data Structure that handles tabular data like spreadsheet, with column and row indices. Usually has similar Data typed values within each column. Logically, the data is stored in a one or two dimensional block. Also permits the handling of high-dimensions through Hierarchial Indexing.



*IMPORTING PANDAS*

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

## *SERIES IN PANDAS*

```python
s=pd.Series([1,2,3,4,5,6,7])
print(s)
print("Values:",s.values)
print("Index:",s.index)
```

```python
#Explicitly Declaring the Indices for it's values
s1=pd.Series([34,65,87,89,66,87],index=['a','b','c','d','e','f'])
print(s1.values)
print(s1.index)
```

```python
#Accessing individual elements
print(s1['a'])
print(s1[['a','b','c']])
```

```python
#Changing the value of the Element
s1['e']=-10
print(s1)
```

```python
#Filtering with boolean array
print(s1[s1>0])
print(s1[s1==89])
```

```python
#Membership operator
print('a' in s1) #Index value
print(34 in s1) #value
```

```python
#Creating Series from dictionaries
students={"Name":"Mr.X","Dept":"CS","Designation":"Asst. Prof"}
s=pd.Series(students)
print(s)
```

```python
#Assigning index values
s2=pd.Series(students,index=["Name","Dept","Address"]) #Address has no values assigned. Hence, returns a NaN
print(s2)
```

```python
#Isnull and notnull() functions
s2.isnull()
```

```python
s2.notnull()
```

```python
#Name Attribute 
s2.name='Staff Details'
s2.index.name="Properties"
print(s2)
```

## *DATAFRAMES IN PANDAS*

```python
#Creating DataFrames from Dictionaries
r={"Rollno":[501,502,503,504,505],"Names":["Mr.A","Mr.B","Mr.C","Mr.D","Mr.E"],"Grades":['A','A+','B','C','A']}
df=pd.DataFrame(r,index=[1,2,3,4,5])
print(df)
```

```python
#The sequence of columns can be explicitly specified
df1=pd.DataFrame(r,columns=["Names","Grades","Rollno"],index=['a','b','c','d','e'])
print(df1)
```

```python
#NaN on non-specified columns
df1=pd.DataFrame(r,columns=["Names","Grades","Rollno","Marks"],index=['a','b','c','d','e'])
print(df1)
```

```python
#Accessing column names
df1.columns
```

```python
#Accessing individual columns
df['Names']
```

```python
#Accessing individual items
df['Names'][3]
```

```python
#Name attribute
df.index.name='S.no'
df.columns.name='Properties'
print(df)
```

```python
#Assigning values to columns
df1['Marks']=np.arange(90,95)
print(df1)
```

```python
#adding a column from a series
s=pd.Series([1,1,0,1,0],index=['a','b','c','d','e'])
df1["Results"]=s
print(df1)
```

```python
#Adding a boolean column
df1["Pass"]=df1["Results"]==1
print(df1)
```

```python
#Deleting a column
del df1["Pass"]
print(df1)
```

```python
#Creating transpose
df1.T
```

```python
#Accessing values
df1.values
```
